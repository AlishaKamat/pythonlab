## Data Visualization

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = sns.countplot( x ='gender',hue='test preparation course', data=df)
ax.set(title="Gender vs test course", xlabel='Gender', ylabel='Total')
plt.show()
```

```python
ax = sns.countplot( x ='gender',hue='race/ethnicity', data=df)
ax.set(title="Gender vs test course", xlabel='Gender', ylabel='Total')
plt.show()
```
